### How do your variables relate to each other? You might look at correlations, scatter plots and/or correlation matrices.
are there big differences between locations? how many persons per trip ?

```python
corr = data.corr()

corr = corr.replace(1, np.nan)
corr
```

The correlation matrix shows the correlation of each value with the others. 

```python
corr.min()
```

The minimum (or highest negative) correlation between the variables for each variable shows that most are not negatively correlated, and that the negative correlations are weak. Meaning that its unlikely that collections with more of one item directly correlate to less of another item. 

```python
corr.max()
```

The max correlation shows that some variables are very closely correlated. An increase in one variable is correlated with an increase in the other. 

```python
PB = data['Other Plastic Bags']
CP = data ['Cups, Plates (Foam)']
correlationplot = pd.concat([PB,CP],axis = 1)

sns.lineplot(data = correlationplot)
```

Taking the two variables with the highest correlation, plotting the lineplots of both, we can see that they do seem to correlate closely - more plastic bags are associated with more foam cups/plates

```python
mp.pyplot.figure(figsize = (30,15), dpi = 300)
sns.barplot(data, x = 'Area', y = 'Total Pounds of Litter Collected')
```

Again, a little hard to read the locations, but this plot shows the amount of trash collected at each location. Some locations had more than one colleciton day (ones with a black std bar), resulting in a higher amount, and others had no collection at all.  

```python
sns.lineplot(data, x = 'Number of Volunteers', y = 'Total Pounds of Litter Collected')
```

This plot shows how many volunteers were present vs the amount of trash collected. Since the line plot is all over the place, it is evident that there is not a high correlation between the two values. Some low number of volunteers collected more trash than high numbers of volunteers.

```python
sns.lineplot(data, x = 'Volunteer Hours', y = 'Total Pounds of Litter Collected')
```

Volunteer hours and total trash collected also don't seem to direclty correlate, although they look better than the number of volunteers. This could be due to some outliers.

### look at range of years and dates and sites? Were some locations only visited once, twice, all in the same year etc.

```python
data[['Month', 'Day', 'Year']] = data.Date.str.split("/", expand = True)

data.groupby('Year')['Total Pounds of Litter Collected'].sum()
```

```python
data.groupby('Year')['Total Pounds of Litter Collected'].count()
```

2022 had the least amount of trash collected, but it had the same number of collection days as 2018. 2016 had the most collection days, but not the most trash collected. 2017 had the highest amount of trash collected, but had only 11 collection days, like 2019 and 2021. 2020 had no collection days.
